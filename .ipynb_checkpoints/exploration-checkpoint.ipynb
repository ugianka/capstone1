```python
from os import listdir
from os.path import isfile, join
import pandas as pd
import math
```

```python
data_path='./cs-train'
onlyfiles = [join(data_path, f) for f in listdir(data_path) if isfile(join(data_path, f))]
```

```python
inpDf=pd.DataFrame()
```

```python
for idx,fname in enumerate(onlyfiles):
    inpDf= inpDf.append(pd.read_json(fname))
```

```python
inpDf
```

```python
#clean the dataFrame by unifyng the columns move data from StreamID and TimesViewed to stream_id and times_viewed

cleanedDf=pd.DataFrame()
for idx,row in inpDf.iterrows():
    bogusStream=row['StreamID']
    bogusTimesViewed=row['TimesViewed']
    
    # check is StreamID is not NaN
    if(bogusStream==bogusStream): 
        row['stream_id']=row['StreamID']
     # check is TimesViewed is not NaN
    if(bogusTimesViewed==bogusTimesViewed):
        row['times_viewed']=row['TimesViewed']
```
